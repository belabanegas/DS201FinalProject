Beginning cleaning by dropping columns

```python
import pandas as pd

url = "https://raw.githubusercontent.com/belabanegas/DS201FinalProject/main/Covid%20Data%20(1).csv"

df = pd.read_csv(url)
df = df.drop (columns= ['DIABETES', 'COPD','HIPERTENSION','PATIENT_TYPE','INMSUPR', 'OTHER_DISEASE', 'RENAL_CHRONIC', 'CLASIFFICATION_FINAL', 'USMER', 'MEDICAL_UNIT', 'ICU'])
```

```python
def deadOrAlive(row):
  if (row['DATE_DIED'] == '9999-99-99'):
    return 'alive'
  else:
    return 'dead'

df['status'] = df.apply(deadOrAlive, axis=1)

df
```

```python
new_df = df[['status', 'PNEUMONIA']]
new_df = new_df.groupby('status').mean()

new_df.plot(kind = 'bar', title = 'Pnemonia vs status')
```

```python
new_df = df[['AGE', 'status']]
new_df = new_df.groupby('status').mean()

new_df.plot(kind = 'bar', title = 'AGE vs status')
```
